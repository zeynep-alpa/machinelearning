```python
import pandas as pd

file_path = 'bank.csv'
data = pd.read_csv(file_path, delimiter=';')

# Veri setinin ilk birkaç satırına göz atarak yapısını kontrol etme
print("Veri setinin ilk birkaç satırı:")
print(data.head())

# Veri setinin yapısını genel olarak anlama
print("\nVeri seti hakkında genel bilgi:")
print(data.info())

# Sayısal sütunlara dair özet istatistikleri görüntüleme
print("\nVeri setinin sayısal sütunları için özet istatistikler:")
print(data.describe())
```

```python
import pandas as pd

# Eksik verileri kontrol et
print("Eksik veri sayısı (her sütun için):")
print(data.isnull().sum())

# Eksik verileri doldurma
for column in data.columns:
    if data[column].dtype == 'object':  # Kategorik sütunlar
        data[column] = data[column].fillna(data[column].mode()[0])
    else:  # Sayısal sütunlar
        data[column] = data[column].fillna(data[column].median())

# Eksik verilerin doldurulduktan sonraki durumunu kontrol eder
print("\nEksik veri sayısı (doldurduktan sonra):")
print(data.isnull().sum())
```

```python
import numpy as np

# Sayısal sütunlardaki aykırı değerleri alt ve üst sınırlara eşitleme
for column in data.select_dtypes(include=[np.number]).columns:
    Q1 = data[column].quantile(0.25)  # 1. çeyrek (25. yüzde)
    Q3 = data[column].quantile(0.75)  # 3. çeyrek (75. yüzde)
    IQR = Q3 - Q1  # Çeyrekler arası mesafe

    # Alt ve üst sınırları hesapla
    lower_limit = Q1 - 1.5 * IQR
    upper_limit = Q3 + 1.5 * IQR

    # Değerleri alt ve üst sınırlara eşitleme
    data[column] = np.clip(data[column], lower_limit, upper_limit)

print("Aykırı değerler IQR yöntemi ile sınırlandırıldı.")
print(data.describe())
```

```python
from sklearn.preprocessing import LabelEncoder

# Kategorik sütunları seçer
categorical_columns = data.select_dtypes(include=['object']).columns

# LabelEncoder ile kategorik sütunları sayısal değerlere dönüştürür
label_encoder = LabelEncoder()
for column in categorical_columns:
    data[column] = label_encoder.fit_transform(data[column])

# Sonuçları kontrol eder
print("Kategorik değişkenler sayısal değerlere dönüştürüldü.")
print(data.head())
```

```python
# Kategorik sütunları one-hot encoding ile dönüştür
data = pd.get_dummies(data, columns=categorical_columns)

# Sonuçları kontrol et
print("Kategorik değişkenler One-Hot Encoding ile sayısal değerlere dönüştürüldü.")
print(data.head())
```
